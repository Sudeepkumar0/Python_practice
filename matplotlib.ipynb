```python
#pie chart

labels = 'Python', 'C++', 'Java'
sizes= [300, 125, 70]
colors = ['r','g','b']
explode = (0.1, 0, 0)  #explode first slide

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct = '%1.1F%%', shadow=True)

plt.axis('equal')
plt.show()
```
